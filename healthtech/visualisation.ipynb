```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv("./data/state_data.csv");
```

```python
def plot_bar_x(df,state):
    diseases = ['Diarrhoeal','Malaria','Respiaratory','Japanese Encephalitis','Viral Hepatitis']
    df = df[["State"]+diseases]
    x = df.loc[df["State"]==state].reset_index(drop=True)
    
    percent_count = []
    for i in range(5):
        percent_count.append(x.at[0,diseases[i]])
    
    index = np.arange(len(diseases))
    plt.bar(index, percent_count)
    plt.xlabel('Diseases', fontsize=5)
    plt.ylabel('Percentage of Death out of total cases', fontsize=10)
    plt.xticks(index, diseases, fontsize=10, rotation=90)
    plt.title('State Wise Visualisation')
    plt.show()
```

```python
state = input("Enter State: ");
```

```python
plot_bar_x(df,state)
```

```python
data_path = './data/Data_Compiled.csv'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
disease_df = pd.read_csv(data_path)
print(type(disease_df))
```

```python
data = disease_df.values
print(type(data))
print(data.shape)
```

```python
disease_df.columns
```

```python
disease_df
```

```python
dataset_andhra = disease_df.loc[1]
disease_andhra = disease_df.loc[0]
```

```python
dataset_andhra
```

```python
andhra_count = dict(dataset_andhra)
andhra_count.pop('State/UTs')
```

```python
andhra_count
```

```python
fig = plt.figure(figsize=(10,6))
plt.pie(andhra_count.values(), labels=andhra_count.keys(), autopct='%1.1f%%',  shadow=True, startangle=90)
plt.title('Disease Count of Andhra Pradesh')
plt.axis('equal')
plt.savefig('./andhra_count.png', dpi=100)
plt.show()
```
